```python
for sentence in ingredients:
    # Parse the ingredient phrase using ingredient_parser
    try:
        parsed = parse_ingredient(sentence, separate_names=True)  # or foundation_foods=True
        name_texts = [n.text for n in parsed.name]
    except Exception as e:
        name_texts = [sentence]

    # For each parsed ingredient name (can be more than one if separated by parser)
    for name in name_texts:
        # Singularize each word in the ingredient name
        singularized = []
        for word in name.split():
            word_proc = word.strip().lower().replace('-', ' ')
            if word_proc:
                result = p.singular_noun(word_proc)
                singularized.append(result if result else word_proc)
        # Join into a single normalized string
        ingredient_str = ' '.join(singularized)
        print("ingredient after singularize:", ingredient_str)

        # Embed the ingredient string
        ing_vec = minilm_model.encode([ingredient_str])[0]

        # Cosine similarity
        dot_products = np.dot(product_embeddings, ing_vec)
        norms = np.linalg.norm(product_embeddings, axis=1) * np.linalg.norm(ing_vec)
        similarities = dot_products / (norms + 1e-8)

        # Find best match
        best_idx = np.argmax(similarities)
        best_sim = similarities[best_idx]
        carbs = carb_values[best_idx]
        matched_product = product_names[best_idx]

        print(f"ING: {ingredient_str}, best_sim: {best_sim:.2f}  Closest product: '{matched_product}' (sim: {best_sim:.2f}, carbs: {carbs})")
```

```python
# filter only one extra word
import re
from IPython.display import display, Markdown

# Items to search and their replacement carb values
search_items = [
    ('sour cream', 3),
    ('parmesan cheese', 4),
    ('whipping cream', 3),
    ('homemade style ice', 0),
    ('lemon', 9),
    ('walnut', 9)
]

# Filter by carbs > 10 once
filtered_df = df[df['carbohydrates_100g'] > 10].copy()

# Add a helper column for word count (case-insensitive cleanup)
filtered_df["word_count"] = filtered_df["product_name"].str.split().str.len()
filtered_df["product_name_lower"] = filtered_df["product_name"].str.lower()

# Display each match separately with no more than 1 extra word
for name, _ in search_items:
    search_words = name.lower().split()
    name_len = len(search_words)

    matches = filtered_df[
        filtered_df["product_name_lower"].str.contains(re.escape(name.lower())) &
        (filtered_df["word_count"] <= name_len + 1)
    ]

    display(Markdown(f"### Matches for `{name}` (≤ 1 extra word)"))
    display(matches.drop(columns=["word_count", "product_name_lower"]))
```
